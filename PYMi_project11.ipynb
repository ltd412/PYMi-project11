## ex11.2 - Đồ thị sin, cos, tan

```python
x = np.linspace(-2*np.pi, 2*np.pi)
y = np.linspace(-1,1)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.legend(["y = sin(x)", "y = cos(x)"])
plt.title("Biểu đồ y = sin(x), y = cos(x)")
```

```python
plt.plot(x, np.tan(x))
plt.legend("y = tan(x)")
plt.title("Biểu đồ y = tan(x)")
```

## ex11_3 - Biểu đồ x = y^2 và x = y

```python
x = np.linspace(-2,2)
```

```python
plt.plot(x**2, x)
plt.plot(x, x)
plt.title(" Đồ thị x = y^2 và x = y")
plt.legend(["x = y^2", "x = y"])
```
